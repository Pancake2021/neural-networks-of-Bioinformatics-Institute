```python
import numpy as np
import matplotlib.pyplot as plt

def poly(x, y, n):
    x0 = x.reshape(x.size, 1)
    q = 0
    X = np.ones((x.size, 1))
    while q <= n:
        cts = np.polyfit(x, y, q)
        Y = X.dot(cts[::-1])
        X = np.hstack((X, x0**(q+1)))
        q += 1
    return Y

# так делал автор видео, наверное ;-ъ
# x0 = x.reshape(x.size, 1)

# X0 = np.ones((x.size, 1))
# cts0 = np.polyfit(x, y, 0)
# Y0 = X0.dot(cts0)

# X1 = np.hstack((X0, x0))
# cts1 = np.polyfit(x, y, 1)
# Y1 = X1.dot(cts1[::-1])

# X2 = np.hstack((X1, x0**2))
# cts2 = np.polyfit(x, y, 2)
# Y2 = X2.dot(cts2[::-1])

# X3 = np.hstack((X2, x0**3))
# cts3 = np.polyfit(x, y, 3)
# Y3 = X3.dot(cts3[::-1])

step = 0.1
start = 0
end = 2 * np.pi

x = np.random.choice(np.arange(start, end + step, step), 35)
x.sort()
y = np.sin(x) + np.random.normal(0, 0.08, x.size) # normal - разброс

x0 = x.reshape(x.size, 1)

X0 = np.ones((x.size, 1))
cts0 = np.polyfit(x, y, 0)
Y0 = X0.dot(cts0)

Y1 = poly(x, y, 1)
Y2 = poly(x, y, 2)
Y3 = poly(x, y, 3)
Y10 = poly(x, y, 10)

plt.scatter(x, y)
plt.plot(x, Y0, label = "degree 0")
plt.plot(x, Y1, label = "degree 1")
plt.plot(x, Y2, label = "degree 2")
plt.plot(x, Y3, label = "degree 3")
plt.plot(x, Y10, label = "degree 10")

plt.legend()
plt.grid()
plt.show()
```
